```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
digitalCoreNames = ["testDC.csv"]
digitalCores = []
for name in digitalCoreNames:
    digitalCores.append(pd.read_csv(name))
for i, digitalCore in enumerate(digitalCores):
    #"Duration(s)","Distance(steps)","Force(N)","Amperage(A)"
    x = (digitalCore["Amperage(A)"])
    fig, ax = plt.subplots(2, 2)
    ax[0, 0].plot(x, (digitalCore["Duration(s)"]))
    ax[0, 1].plot(x, (digitalCore["Distance(steps)"]))
    ax[1, 0].plot(x, (digitalCore["Force(N)"]))
    ax[1, 1].plot(x, (digitalCore["Amperage(A)"]))
    fig.savefig("fig"+str(i)+".png")
```
